```python
import pandas as pd
import numpy as np

# Dataset
data = {
    'Distance': [10, 50, 100, 150, 200],
    'Weight': [5, 10, 15, 20, 25],
    'Fuel_Price': [80, 82, 87, 90, 95],
    'cost': [500, 1000, 1500, 2000, 2500]
}

df = pd.DataFrame(data)

# Input & Output
X = df[['Distance','Weight','Fuel_Price']]
y = df['cost']   # ✅ fixed

# Model
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X,y)

# Prediction input (correct way)
input_data = pd.DataFrame({
    'Distance': [120],
    'Weight': [15],
    'Fuel_Price': [88]
})

prediction = model.predict(input_data)

print("Prediction cost:", prediction)
import pickle
pickle.dump(model,open('model.pkl','wb'))
```
